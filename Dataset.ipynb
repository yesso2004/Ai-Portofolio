```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("WineQT.csv") # read the csv file from excel
dataset.head(10) # Read the first 10 lines and showcase the table
```

```python
x = dataset.iloc[:,:-2].values # Set the x variable to all the coulmuns except the last 2
y = dataset.iloc[:,-2].values # Set y variable to the before last columns (quality)
```

```python
from sklearn.model_selection import train_test_split # import training and testing model

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) # split the dataset to training and testing subsets 
# x represents the input data and y represents that data to be predicted 
# test_size represent the ratio that we want training and testing and in this case its 75% training and 25% testing 
# random_state is set to 0 to ensure the data  split will be same  
```

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler() # makes the data easier for the machine learnig algorithm to understand by adjusting the average of each column becomes zero 
x_train = scaler.fit_transform(x_train) # calculates the mean of the data and 
x_test = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

print("Predicated vs labels")
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```
